```python
# División por cero
try:
    numero = 7 / 0
except ZeroDivisionError:
    print("Error: No se puede dividir por cero. Por favor, usa un divisor distinto de 0.")
```

```python
# Índice fuera de rango
lista = [4, 7, 30, 23, 5]

try:
    print(lista[10])
except IndexError:
    print("Error: Índice fuera de rango. La lista tiene", len(lista), "elementos.")
```

```python
# Clave inexistente en diccionario
paises = {"españa": "español", "eeuu": "inglés", "italia": "italiano"}

try:
    print(paises["alemania"])
except KeyError:
    print("Error: La clave 'alemania' no existe en el diccionario. Usa claves válidas:", list(paises.keys()))
```

```python
# Error de tipo
try:
    Resultado = "2" + 10
except TypeError:
    print("Error: No se pueden combinar cadenas y números directamente. Convierte los tipos antes de operar.")
```

```python
# Claves de entrada de usuario
datos = {
  "nombre": "Juan",
  "edad": 30,
  "ciudad": "Madrid"
}

clave = input("Introduce una clave para obtener su valor: ")

try:
    print(f"El valor asociado a '{clave}' es: {datos[clave]}")
except KeyError:
    print(f"Error: La clave '{clave}' no existe. Introduce una de las siguientes claves:", list(datos.keys()))
```

```python
# Conversión y división
def calcular(numero1, numero2):
    return (int(numero1) ** 2) / int(numero2)

n1 = input("Introduce el primer número: ")
n2 = input("Introduce el segundo número: ")

try:
    resultado = calcular(n1, n2)
    print(f"El resultado del cálculo es: {resultado}")
except ValueError:
    print("Error: Debes introducir valores numéricos válidos.")
except ZeroDivisionError:
    print("Error: No se puede dividir por cero. Introduce un segundo número distinto de 0.")
```

```python
# Función para agregar sin repetir
def agregar_sin_repetidos(lista, elemento):
    try:
        if elemento in lista:
            raise ValueError(f"Imposible añadir elementos duplicados => {elemento}")
        lista.append(elemento)
        print(f"Elemento '{elemento}' añadido correctamente.")
    except ValueError as e:
        print("Error:", e)

# Ejemplo de uso
mi_lista = [1, 2, 3]
agregar_sin_repetidos(mi_lista, 2)  # Elemento duplicado
agregar_sin_repetidos(mi_lista, 4)  # Elemento nuevo
```
