```python
# Función para calcular el área de un rectángulo
def area_rectangulo(base, altura):
    """
    Calcula el área de un rectángulo.
    
    Parámetros:
    base (float): base del rectángulo
    altura (float): altura del rectángulo
    
    Retorna:
    float: área del rectángulo
    """
    return base * altura

# Solicitar datos al usuario
base = float(input("Introduce la base del rectángulo: "))
altura = float(input("Introduce la altura del rectángulo: "))

# Calcular e imprimir el área
area = area_rectangulo(base, altura)
print(f"El área del rectángulo es: {area}")
```

```python
import math

def area_circulo(radio):
    """
    Calcula el área de un círculo.
    
    Parámetros:
    radio (float): radio del círculo
    
    Retorna:
    float: área del círculo
    """
    return math.pi * radio ** 2

# Solicitar radio al usuario
radio = float(input("Introduce el radio del círculo: "))

# Calcular e imprimir el área
area = area_circulo(radio)
print(f"El área del círculo es: {area}")
```

```python
def lee_numero():
    """
    Solicita un número al usuario y lo retorna.
    """
    return float(input("Introduce un número: "))

def mayor(a, b, c):
    """
    Devuelve el mayor de tres números.
    """
    return max(a, b, c)

# Solicitar tres números
num1 = lee_numero()
num2 = lee_numero()
num3 = lee_numero()

# Encontrar e imprimir el mayor
print(f"El número mayor es: {mayor(num1, num2, num3)}")
```

```python
def imc(peso, altura):
    """
    Calcula el índice de masa corporal y su clasificación.
    """
    valor_imc = peso / (altura ** 2)
    if valor_imc < 18.50:
        clasificacion = "Bajo peso"
    elif valor_imc < 25.00:
        clasificacion = "Normal"
    elif valor_imc < 30.00:
        clasificacion = "Sobrepeso"
    else:
        clasificacion = "Obesidad"
    return valor_imc, clasificacion

# Solicitar peso y altura al usuario
peso = float(input("Introduce tu peso en kg: "))
altura = float(input("Introduce tu altura en metros (ej: 1.75): "))

# Calcular e imprimir IMC y clasificación
valor, categoria = imc(peso, altura)
print(f"Tu IMC es {valor:.2f}, clasificación: {categoria}")
```

```python
def pedir_contraseña():
    return input("Crea una contraseña: ")

def validar_login(contraseña_guardada):
    intentos = 0
    while intentos < 5:
        intento = input("Introduce tu contraseña para loguearte: ")
        if intento == contraseña_guardada:
            print("Login correcto")
            return True
        else:
            intentos += 1
            print(f"Login fallido. Intento {intentos}")
    print("Se han agotado los intentos. Reiniciando...")
    return False

# Programa principal
while True:
    contraseña = pedir_contraseña()
    if validar_login(contraseña):
        break
```

```python
def intermedio(num1, num2):
    """
    Calcula el punto intermedio entre dos números.
    """
    return (num1 + num2) / 2

# Solicitar números al usuario
n1 = float(input("Introduce el primer número: "))
n2 = float(input("Introduce el segundo número: "))

# Calcular e imprimir el punto intermedio
punto = intermedio(n1, n2)
print(f"El punto intermedio entre {n1} y {n2} es: {punto}")
```

```python
def recortar(numero, limite_inferior, limite_superior):
    """
    Recorta un número según los límites dados.
    """
    if numero < limite_inferior:
        return limite_inferior
    elif numero > limite_superior:
        return limite_superior
    else:
        return numero

# Solicitar datos al usuario
numero = float(input("Introduce el número a recortar: "))
limite_inf = float(input("Introduce el límite inferior: "))
limite_sup = float(input("Introduce el límite superior: "))

# Aplicar recorte y mostrar resultado
numero_recortado = recortar(numero, limite_inf, limite_sup)
print(f"El número recortado es: {numero_recortado}")
```

```python
def separar(lista):
    """
    Separa los números de una lista en pares e impares y los ordena.
    """
    pares = sorted([x for x in lista if x % 2 == 0])
    impares = sorted([x for x in lista if x % 2 != 0])
    return pares, impares

# Ejemplo de lista
numeros = [-12, 84, 13, 20, -33, 101, 9]

# Separar e imprimir
pares, impares = separar(numeros)
print(pares)
print(impares)
```
