```python
# ==========================================
# Ejercicio 1: Listas y Tuplas
# ==========================================
mi_lista = ["manzana", "banana", "cereza"]
mi_tupla = ("perro", "gato", "conejo")

print(mi_lista)
print(mi_tupla)

print(mi_lista[1])        # Segundo elemento de la lista
print(mi_tupla[-2])       # Penúltimo elemento de la tupla

mi_lista[0] = "kiwi"
print(mi_lista)

print(len(mi_lista), len(mi_tupla))

print("banana" in mi_lista)
print("gato" in mi_tupla)

mi_lista.append("naranja")
print(mi_lista)

mi_lista.clear()
print(mi_lista)

# ==========================================
# Ejercicio 2: Sets y Diccionarios
# ==========================================
mi_set = {"rojo", "verde", "azul"}
mi_dic = {"nombre": "Ana", "edad": 25, "ciudad": "Madrid"}

print(mi_set)
print(mi_dic)

print(mi_dic["nombre"])

mi_dic["edad"] = 26
print(mi_dic)

print(len(mi_set), len(mi_dic))

print("rojo" in mi_set)
print("edad" in mi_dic)

mi_set.add("amarillo")
mi_dic["pais"] = "España"
print(mi_set, mi_dic)

mi_set.clear()
mi_dic.clear()
print(mi_set, mi_dic)

# ==========================================
# Ejercicio 3: Suma de elementos en lista
# ==========================================
numeros = [1.5, 2.3, 4.2]
sumatorio = sum(numeros)
print(sumatorio)

# ==========================================
# Ejercicio 4: Media aritmética dinámica
# ==========================================
media = sum(numeros) / len(numeros)
print(media)

# ==========================================
# Ejercicio 5: Lista de la compra
# ==========================================
productos = ["pan", "leche", "huevos", "queso", "manzana", "banana"]
print(productos[1], productos[-1])

productos[2] = "yogur"
print(productos)

productos[0], productos[-1] = productos[-1], productos[0]
print(productos)

print(productos[:3])
print(productos[-3:])
print(productos[::2])

# ==========================================
# Ejercicio 6: Formateo de cadenas invertidas
# ==========================================
cadena = "zereP solraC, 5891"
cadena_invertida = cadena[::-1]  # Resultado: '1985 Carlos Pérez'

# Separar año y nombre basándonos en que los 4 primeros caracteres son el año
anio = cadena_invertida[:4]
nombre_apellido = cadena_invertida[5:].strip()  # Saltar el espacio después del año
print(f"Empleado {nombre_apellido} nacido en el año {anio}")

# ==========================================
# Ejercicio 7: Operaciones con conjuntos
# ==========================================
mamiferos = {"Gato", "León", "Caballo", "Ballena", "Tortuga"}
peces = {"Ballena", "Sardina"}

mamiferos.discard("Tortuga")
peces.add("Pez Payaso")

print(mamiferos, peces)
print(mamiferos & peces)

# ==========================================
# Ejercicio 8: Comparación de alumnos
# ==========================================
alumnos_python = {"Ana", "Luis", "Pedro"}
alumnos_ia = {"Luis", "Marta", "Pedro"}

ambos = list(alumnos_python & alumnos_ia)
solo_python = list(alumnos_python - alumnos_ia)
solo_ia = list(alumnos_ia - alumnos_python)

print(ambos, solo_python, solo_ia)

# ==========================================
# Ejercicio 9: Depurando duplicados
# ==========================================
lista_con_duplicados = ["manzana", "pera", "manzana", "naranja"]
lista_sin_duplicados = list(dict.fromkeys(lista_con_duplicados))
print(lista_sin_duplicados)

# ==========================================
# Ejercicio 10: Gestión de tareas con diccionario
# ==========================================
tareas = {"pendientes": [], "en_progreso": [], "completadas": []}

nueva_tarea = "Comprar leche"
estado = "pendientes"

if estado in tareas:
    tareas[estado].append(nueva_tarea)

print(tareas)
```
