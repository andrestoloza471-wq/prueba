```python
# Lectura de números
num1 = int(input("Introduce el primer número: "))
if num1 == 0:
    print("Error: el primer número no puede ser 0.")
else:
    num2 = int(input("Introduce el segundo número: "))
    num3 = int(input("Introduce el tercer número: "))
    
    # Verificar orden ascendente
    if num1 < num2 < num3:
        print("Los números están en orden ascendente.")
    else:
        print("Los números NO están en orden ascendente.")
```

```python
# Lectura de números en una lista
numeros = []
numeros.append(int(input("Introduce el primer número: ")))
numeros.append(int(input("Introduce el segundo número: ")))
numeros.append(int(input("Introduce el tercer número: ")))

# Validación
if numeros[0] == 0:
    print("Error: el primer número no puede ser 0.")
else:
    if numeros == sorted(numeros):
        print("Los números están en orden ascendente.")
    else:
        print("Los números NO están en orden ascendente.")
```

```python
# Solicitar el valor de N
N = int(input("Introduce el valor de N: "))

sumatorio = 0
for i in range(N):
    sumatorio += i

print(f"El sumatorio de los primeros {N} números es: {sumatorio}")
```

```python
contador_a = 0

while True:
    letra = input("Introduce una letra ('.' para terminar): ")
    if letra == '.':
        break
    if letra.lower() == 'a':
        contador_a += 1

print(f"Total de letras 'a' introducidas: {contador_a}")
```

```python
contador_a = 0

while True:
    palabra = input("Introduce una palabra ('.' para terminar): ")
    if palabra == '.':
        break
    contador_a += palabra.lower().count('a')

print(f"Total de apariciones de la letra 'a': {contador_a}")
```

```python
sumatorio = 0

for numero in range(1000, 72, -1):
    if numero % 2 == 0:
        sumatorio += numero

print(f"El sumatorio de los números pares es: {sumatorio}")
```

```python
cantidad = int(input("¿Cuántos valores deseas introducir?: "))
valores = []

for i in range(cantidad):
    numero = float(input(f"Introduce el número {i+1}: "))
    valores.append(numero)

# Ordenar la lista
valores_ordenados = sorted(valores)

# Cálculos
promedio = sum(valores) / len(valores)
maximo = max(valores)
minimo = min(valores)
pos_max = valores.index(maximo)
pos_min = valores.index(minimo)

print("Lista ordenada:", valores_ordenados)
print("Promedio:", promedio)
print("Máximo:", maximo, "en la posición", pos_max)
print("Mínimo:", minimo, "en la posición", pos_min)
```

```python
compras = []
ventas = []
alquileres = []

while True:
    print("""
MENU COMPRA-VENTA
1. Comprar
2. Vender
3. Alquilar
4. Mostrar listas de datos
5. Salir
""")
    opcion = input("Elige una opción: ")

    if opcion == '1':
        item = input("¿Qué deseas comprar?: ")
        compras.append(item)
    elif opcion == '2':
        item = input("¿Qué deseas vender?: ")
        ventas.append(item)
    elif opcion == '3':
        item = input("¿Qué deseas alquilar?: ")
        alquileres.append(item)
    elif opcion == '4':
        print("Compras:", compras)
        print("Ventas:", ventas)
        print("Alquileres:", alquileres)
    elif opcion == '5':
        print("Saliendo del programa...")
        break
    else:
        print("Opción no válida.")
```

```python
while True:
    contraseña_guardada = input("Crea una contraseña: ")
    intentos = 0

    while intentos < 5:
        contraseña_ingresada = input("Introduce la contraseña para iniciar sesión: ")
        if contraseña_ingresada == contraseña_guardada:
            print("Login correcto")
            break
        else:
            intentos += 1
            print(f"Login fallido. Intento {intentos}")

    if intentos < 5:
        break
    else:
        print("Se alcanzó el máximo de intentos. Reiniciando programa...")
```
