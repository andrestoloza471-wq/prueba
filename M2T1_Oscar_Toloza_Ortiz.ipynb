```python
# Ejercicio 1: Longitud de una cadena
mensaje = "Hola, bienvenidos al curso de Python"

# Obtener el número total de caracteres
longitud = len(mensaje)

# Imprimir el resultado
print("La longitud de la cadena es:", longitud)
```

```python
# Ejercicio 2: Índices de cadenas
texto = "Python"

# Primera letra
primera_letra = texto[0]
print("Primera letra:", primera_letra)

# Última letra usando índice negativo
ultima_letra = texto[-1]
print("Última letra:", ultima_letra)
```

```python
# Ejercicio 3: Extraer partes de una dirección
direccion = "Calle Falsa, 123, Springfield"

# Extraer nombre de la calle
calle = direccion.split(",")[0]
print("Calle:", calle)

# Extraer número
numero = direccion.split(",")[1].strip()  # .strip() elimina espacios
print("Número:", numero)

# Extraer ciudad
ciudad = direccion.split(",")[2].strip()
print("Ciudad:", ciudad)
```

```python
# Ejercicio 4: Identificador único
nombre = "Cristian"
apellido = "Rodriguez"
anio_nacimiento = "1989"

# Extraer partes necesarias
id_nombre = nombre[:3]       # primeras 3 letras
id_apellido = apellido[:3]   # primeras 3 letras
id_anio = anio_nacimiento[-2:]  # últimos 2 dígitos

# Concatenar para formar identificador
identificador = id_nombre + id_apellido + id_anio
print("Identificador generado:", identificador)
```

```python
# Ejercicio 5: Inmutabilidad
palabra = "robot"

# Intento de modificar directamente (esto generará un error)
# palabra[0] = "R"  # <-- Esto daría TypeError

# Solución usando slicing y concatenación
palabra_modificada = "R" + palabra[1:]
print("Palabra modificada:", palabra_modificada)
```

```python
# Ejercicio 6: Analizador de palabras
texto = "Python es POTENTE. Aprender python es útil. PYTHON abre puertas."
objetivo = "Python"

# Convertir todo a minúsculas para búsqueda insensible a mayúsculas
texto_lower = texto.lower()
objetivo_lower = objetivo.lower()

# Contar apariciones exactas como palabra
palabras = texto_lower.split()
conteo_exacto = palabras.count(objetivo_lower)

# Contar apariciones como subcadena
conteo_subcadena = texto_lower.count(objetivo_lower)

print("Apariciones exactas:", conteo_exacto)
print("Apariciones como subcadena:", conteo_subcadena)

# Generar versión resaltada del texto
resaltado = texto.replace("Python", "PYTHON").replace("python", "PYTHON").replace("PYTHON", "PYTHON")
print("Texto resaltado:", resaltado)
```
