```python

# ==========================================
# Ejercicio 1: Tipos de datos
# ==========================================
# Entero
entero = 10
# Flotante
flotante = 3.14
# String
cadena = "Hola"
# Booleano
booleano = True

# ==========================================
# Ejercicio 2: Operaciones con strings
# ==========================================
nombre = "Juan"
apellido = "Pérez"

# Concatenación
concat = nombre + apellido
concat_punto = nombre + apellido + "."
repetir = nombre * 3

# ==========================================
# Ejercicio 3: Slicing
# ==========================================
nombreCompleto = "Juan Pérez"
nombre_extraido = nombreCompleto[0:4]  # 'Juan'

# ==========================================
# Ejercicio 4: Media aritmética
# ==========================================
nota1, nota2, nota3, nota4, nota5 = 6, 7, 5, 8, 6
media = (nota1 + nota2 + nota3 + nota4 + nota5) / 5
print(f"La nota media es {media}")

# ==========================================
# Ejercicio 5: Redondeo
# ==========================================
resultado = (365 / 12) * 14.7
resultado_redondeado = round(resultado, 2)
print(resultado_redondeado)

# ==========================================
# Ejercicio 6: Operadores lógicos
# ==========================================
username = "Pedro"
password = "Python"

usuario_valido = 3 <= len(username) <= 10
password_valido = password == "Tokio" or password == "Python"
print(usuario_valido, password_valido)

# ==========================================
# Ejercicio 7: Operadores de asignación
# ==========================================
a = 10
b = 5
a += b  # a = 15
a -= b  # a = 10
a *= b  # a = 50
a /= b  # a = 10.0

# ==========================================
# Ejercicio 8: Función input() con comprobación de tipo
# ==========================================
# En entornos sin soporte de input(), se usan valores predefinidos para test

def solicitar_dato(tipo_esperado, valor_predefinido):
    try:
        if tipo_esperado == int:
            dato_convertido = int(valor_predefinido)
        elif tipo_esperado == float:
            dato_convertido = float(valor_predefinido)
        elif tipo_esperado == str:
            dato_convertido = str(valor_predefinido)
        else:
            print("Tipo no soportado.")
            return None

        if type(dato_convertido) == tipo_esperado:
            print(f"¡Dato correcto! Tipo: {type(dato_convertido)}")
            return dato_convertido
    except ValueError:
        print(f"Error: No se puede convertir el dato a {tipo_esperado}.")

# Test cases
edad = solicitar_dato(int, 25)
altura_test = solicitar_dato(float, 1.75)
nombre_test = solicitar_dato(str, "Ana")

# ==========================================
# Ejercicio 9: Formateo de números
# ==========================================
numero_entero = 42
numero_formateado = f"{numero_entero:05d}"  # Rellenado con ceros

numero_flotante = 123.456789
numero_flotante_formateado = f"{numero_flotante:05.3f}"

# ==========================================
# Ejercicio 10: Formateo de strings con valores predefinidos (sin input)
# ==========================================
altura = 1.80  # valor predefinido
peso = 80.135  # valor predefinido

# Variantes de format
print("La altura es de {} metros y el peso es de {} KG".format(altura, peso))
print(f"La altura es de {altura:.2f} metros y el peso es de {peso:.3f} KG")
print("La altura es de {h:.2f} metros y el peso es de {p:.3f} KG".format(h=altura, p=peso))
```
