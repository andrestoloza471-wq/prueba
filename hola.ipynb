```python
# Solicitar datos al usuario
dato_entero = int(input("Ingresa un número entero: "))
dato_flotante = float(input("Ingresa un número flotante: "))
dato_string = input("Ingresa un texto: ")

# Verificar el tipo de cada dato
print(f"Tipo de dato del número entero: {type(dato_entero)}")
print(f"Tipo de dato del número flotante: {type(dato_flotante)}")
print(f"Tipo de dato del texto: {type(dato_string)}")
```
